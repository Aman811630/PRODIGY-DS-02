## Data Transforming and Cleaning.

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# droping those where Embarked is null
df = df.drop(columns='Cabin').dropna(subset=['Embarked'])
```

```python
df.info()
```

```python
df
```

## Doing some Feature engineering 

```python
# rounds each value in the 'Age' column to the nearest integer.
df['Age'] = df['Age'].round()

# maps the value 1 to 'Upper', 2 to 'Middle', and 3 to 'Lower'.
df['Pclass'] = df['Pclass'].map({1:'Upper', 2:'Middle', 3:'Lower'})

# maps the value 1 to 'Upper', 2 to 'Middle', and 3 to 'Lower'.
df['Embarked'] = df['Embarked'].map({1:'Upper', 2:'Middle', 3:'Lower'})

#  maps the value 0 to 'Survived' and the value 1 to 'Not Survived'.
df['Survived'] = df['Survived'].map({0:'Survived', 1:'Not Survived'})

# converts the first character of each word to uppercase and the rest to lowercase.
df['Sex'] = df['Sex'].str.title()

categorical_columns = ['Sex', 'Parch', 'SibSp', 'Pclass', 'Embarked', 'Survived']
df[categorical_columns] = df[categorical_columns].astype('category')
```

```python
cols = [
    'Name', 'Sex', 'Age', 'Parch', 'SibSp',
    'Ticket', 'Pclass', 'Embarked', 'Fare',
    'Survived'
]

df = df[cols]
```

### EDA and Data Exploration / Visulization

```python
# ploting the countplot to know the survived Not survived Num
sns.countplot(data=df, x='Survived')

plt.title('Suvived v/s Not Survived')
plt.show();
```

```python
# ploting subplots of Boxplots on Distribution of AGE and Fare.

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))

sns.boxplot(data=df, x='Age', y='Survived', ax=ax1, palette='muted')
ax1.set_xlabel('Age', fontsize=12)
ax1.set_ylabel('Survived', fontsize=12)
ax1.set_title('Distribution of Age')

sns.boxplot(data=df, x='Fare', y='Survived', ax=ax2, palette='muted')
ax2.set_xlabel('Fare', fontsize=14)
ax2.set_ylabel('Survived', fontsize=14)
ax2.set_title('Distribution of Fare')

plt.tight_layout()
plt.show()
```

```python
# # ploting subplots of pairplots on Distribution of AGE and Fare.

sns.pairplot(data=df, hue='Survived', aspect=2);
```

```python
# ploting subplots of countplot on Distribution of AGE and Fare.


cat_cols = ['Sex', 'Pclass', 'Parch', 'SibSp']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for feature, ax in zip(cat_cols, axes.flatten()):
    sns.countplot(data=df, x=feature, hue='Survived', ax=ax)
    ax.set_title(f'Count of {feature}')
    ax.set_xlabel('')
    ax.set_ylabel('')

plt.tight_layout()
plt.show();
```
